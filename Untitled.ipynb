```python
from ultralytics import YOLO
import cv2
from ultralytics.solutions import object_counter
# Define ROI coordinates (adjust as needed)
x1 = 100
y1 = 50
x2 = 300
y2 = 400
line_points = [(102, 180),(108,108), (180, 180),(180, 180)]
# Load the YOLO model
model = YOLO("best.pt")
counter = object_counter.ObjectCounter()
counter.set_args(view_img=True,
                 reg_pts=line_points,
                 classes_names=model.names,
                 draw_tracks=True)
classes_to_count = [0,2,18]
# Open your laptop's camera
cap = cv2.VideoCapture(0)
assert cap.isOpened(), "Error accessing webcam"

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Extract ROI
    roi = frame[y1:y2, x1:x2]

    # Run object detection on ROI
    results = model(roi)

    # Visualize results (modify to consider ROI)
    # results.show()

    # Draw ROI on the original frame (optional)
    cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('Frame', frame)
    
    while cap.isOpened():
        success, im0 = cap.read()
        if not success:
            print("Video frame is empty or video processing has been successfully completed.")
            break
        tracks = model.track(im0, persist=True, show=False,
                             classes=classes_to_count)

        im0 = counter.start_counting(im0, tracks)
    # Press 'q' to quit
    if cv2.waitKey(1) == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
from ultralytics.solutions import object_counter
import cv2

# Initialize YOLO model
model = YOLO("best.pt")


# Set up your video source
cap = cv2.VideoCapture(0)
classes_to_count = [0, 2,18] 
# Initialize Object Counter for your line
line_pedestrian_1 =  [(100, 100), (600, 100), (600, 360), (100, 360)]  # Define as per your need
counter = object_counter.ObjectCounter()
counter.set_args(view_img=True, reg_pts=line_pedestrian_1, classes_names=model.names)

while cap.isOpened():
    ret, im0 = cap.read()
    if not ret:
        break

    # Track objects
    results = model.track(im0,persist=True, show=False, classes=classes_to_count)

    # Start counting based on tracks and your line
    im0 = counter.start_counting(im0, results)

    # Display
    cv2.imshow("Count", im0)
    if cv2.waitKey(1) == ord('q'):  # press q to quit
        break

cap.release()
cv2.destroyAllWindows()
```
